```python
s = "abcdaabccccccceeeeeeee"

# Which char is present max number of times 
max_char = ""
max_value = 0
for char in s:
    char_count = s.count(char) 
    if char_count >= max_value:
        max_value,max_char = char_count,char
        
print(max_char,max_value)
```
