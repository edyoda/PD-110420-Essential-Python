```python
Generate a 8 char long password 
    - 1 Uppercase
    - 1 Lowercase 
    - 1 Special char 
    - 5 digits 
```

```python
print(ord('A'),ord('Z'))
```

```python
def gen_password(k=8):
    special_chars =  ['!','@','#','&','$']
    upper = chr(random.randint(ord('A'),ord('Z')))
    lower = chr(random.randint(ord('a'),ord('z')))
    char = random.choice(special_chars)
    digits = str(random.randint(10000,99999))
    
    fixed_password = upper+lower+char+digits
#     print(fixed_password)
    temp = random.sample(fixed_password,k)
    password = ("").join(temp)
    return password
    
print(gen_password())
```
